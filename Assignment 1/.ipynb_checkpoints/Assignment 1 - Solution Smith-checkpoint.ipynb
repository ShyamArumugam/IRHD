## Task 4
***

Show that the covariance matrix $C$ of any random variable X $\in$ $\mathbb{R}^p$ is symmetric positive semidefinite, i.e. $C = C^⊤$ and $x^⊤ Cx \ge 0$ for any covariance matrix $C \in \mathbb{R}^p×p$ and any $x \in \mathbb{R}^p$.

### Solution

The covariance matrix of a random variable **X** $\textbf{$\in$ $\mathbb{R}^p$}$ with mean **m** is defined via:

\begin{align}
C = E[(X − m)(X − m)^T]
\end{align}


The $(i, j)^{th}$ element of this covariance matrix $C$ is given by

\begin{align}
C_{ij} = E[(X_i − m_i)(X_j − m_j)] = \sigma_{ij}
\end{align}

The diagonal entries of this covariance matrix $C$ are the variances of the components of the random variable X, i.e.,

\begin{align}
C_{ii} = E[(X_i − m_i)^2] = \sigma^2_i.
\end{align}

This covariance matrix $C$ is symmetric, i.e., $C = C^T$ because :

\begin{align}
C_{ij} = \sigma_{ij} = \sigma_{ji} = C_{ji}
\end{align}

The covariance matrix $C$ is positive semidefinite, i.e., for **x** $\textbf{$\in$ $\mathbb{R}^p$}$:

\begin{align}
E[\thinspace[(X − m)^T x]^2\thinspace] = E[\thinspace[(X − m)^T x]^T[(X − m)^T x]\thinspace] \ge 0 \\ \\
E[x^T(X − m)(X − m)^T x] \ge 0 , \enspace x \in \mathbb{R}^p \\ \\
x^T Cx \ge 0, \enspace \enspace x \in \mathbb{R}^p \\ \\
\end{align}
